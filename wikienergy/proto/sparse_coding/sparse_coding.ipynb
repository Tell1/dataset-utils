```python
import pandas as pd
import numpy as np
from multiprocessing import Pool
from sklearn.decomposition import NMF

from matplotlib import rcParams
rcParams['figure.figsize'] = (12,7)
import matplotlib.pyplot as plt

#query data by id and categorize it
categories = ['hvac', 'refrigerator', 'waterheater', 'inside_lights', 'outside_lights', 'livingroom', 'kitchen', 'washer_dryer', 'misc'];
category_map = {
'use': ['use'],
'hvac': ['air1','air2','air3','airwindowunit1','furnace1','furnace2','heater1','housefan1'],
'refrigerator': ['refrigerator1','refrigerator2','freezer1'],
'waterheater': ['waterheater1','waterheater2'],
'livingroom': ['livingroom1','livingroom2'],
'inside_lights': ['bedroom1','bedroom2','bedroom3','bedroom4','bedroom5','bathroom1','bathroom2','diningroom1','diningroom2','lights_plugs1','lights_plugs2','lights_plugs3','lights_plugs4','lights_plugs5','lights_plugs6', ],
'outside_lights': ['outsidelights_plugs1','outsidelights_plugs2'],
'kitchen': ['kitchen1','kitchen2','kitchenapp1','kitchenapp2','oven1','oven2','disposal1','icemaker1','microwave1','dishwasher1','range1','venthood1','winecooler1'],
'washer_dryer': ['clotheswasher1','clotheswasher_dryg1', 'drye1','dryg1'],
'misc': ['unknown1','unknown2','unknown3','unknown4','aquarium1','car1','garage1','garage2','jacuzzi1','office1','pool1','pool2','poollight1','poolpump1','pump1','security1','shed1','sprinkler1','subpanel1','subpanel2','utilityroom1']}

def categorize_data(df):
    categorized = {}
    for category,columns in category_map.iteritems():
        categorized[category] = np.zeros(len(df))
        for column in columns:
            categorized[category] += df[column].astype(float)
    return categorized

def read_pecan_street_sets(path, week=False, n=100):
    days = (7 if week else 1)
    curated_data = pd.read_csv(path, parse_dates=[2], nrows=4*24*days*n).fillna(value=0)
    sets = []
    by_dataid = curated_data.groupby('dataid')
    for data_id,house_data in by_dataid:
        house_data = house_data.set_index('utc_15min')
        if week:
            by_time = house_data.groupby(lambda x: x.week)
        else:
            by_time = house_data.groupby(lambda x: x.day)
        for time, time_data in by_time:
            if len(time_data) == 24*4*days:
                sets.append(categorize_data(time_data))
    return sets

# organize 
def get_training_data(sets):
    training_data = []
    for category in categories:
        c = []
        for set in sets:
            c.append(set[category])
        training_data.append(np.array(c))
    return training_data

# use scikit to learn a dictionary in which the data is sparse
def learn_dictionary(training_data, n_atoms):
    from sklearn.decomposition import DictionaryLearning
    category_dico = []
    for i in range(len(categories)):
        dico = DictionaryLearning(n_atoms)
        dico.fit(training_data[i])
        category_dico.append(dico.components_)
    return np.concatenate(category_dico.values())

def get_sparse_coder(training_data, n_atoms=12):
    from sklearn.decomposition import SparseCoder
    dictionary = learn_dictionary(training_data, n_atoms)
    return SparseCoder(dictionary=dictionary)

def disaggregate(sparse_coder, series):
    set = {}
    dictionary = sparse_coder.components_
    A = sparse_coder.transform(series)
    n_atoms = len(dictionary)/len(categories)
    use = np.zeros(len(series))
    for i,category in enumerate(categories):
        sum = np.zeros(len(series))
        for j in range(n_atoms):
            k = n_atoms*i + j
            sum += A[k]*dictionary[k]
        set[category] = sum
        use += sum
    set['use'] = use
    return set
```

```python
# learn a single category
def learn_nn_dictionary_category(job_data):
    category_data,n_atoms = job_data
    dico = NMF(n_components=n_atoms, sparseness='data')
    dico.fit(category_data)
    return dico.components_

def learn_nn_dictionary(training_data, n_atoms, n_jobs=1):
    p = Pool(n_jobs)
    dicos = p.map(learn_nn_dictionary_category, [(category_data, n_atoms) for category_data in training_data])
    return np.concatenate(dicos)

def get_nn_sparse_coder(training_data, n_atoms=12, n_jobs=1):
    from sklearn.decomposition import NMF
    dictionary = learn_nn_dictionary(training_data, n_atoms, n_jobs)
    coder = NMF(n_components=n_atoms * len(categories))
    coder.components_ = dictionary
    return coder

def nn_disaggregate(sparse_coder, series):
    set = {}
    dictionary = sparse_coder.components_
    A = sparse_coder.transform(series).transpose()
    n_atoms = len(dictionary)/len(categories)
    use = np.zeros(len(series))
    for i,category in enumerate(categories):
        sum = np.zeros(len(series))
        for j in range(n_atoms):
            k = n_atoms*i + j
            sum += A[k]*dictionary[k]
        set[category] = sum
        use += sum
    set['use'] = use
    return set
```
